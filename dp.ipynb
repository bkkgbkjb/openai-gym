```python
Values = List[float]
# values = [float(i) * 100 for i in range(14)]
values = np.random.random(14)

delta = 0.001
gamma = 1
```

```python
def policy_evaluation(policy: Policy, dynamic: Dynamic) -> Values:
    # temp_val = [i for i in range(len(policy))]
    temp_val = np.random.random(len(policy))
    while True:
        _delta = 0.0
        for (i, v) in enumerate(temp_val):
            old_v = v

            temp_val[i] = np.sum(
                list(
                    map(
                        lambda actprob: actprob[1]
                        * np.sum(
                            list(
                                map(
                                    lambda newStateProbReward: newStateProbReward[1][0]
                                    * (
                                        newStateProbReward[1][1]
                                        + gamma
                                        * (
                                            temp_val[newStateProbReward[0] - 1]
                                            if newStateProbReward[0] != -1
                                            else 0
                                        )
                                    ),
                                    dynamic(i + 1, actprob[0]).items(),
                                )
                            )
                        ),
                        policy[i].items(),
                    )
                )
            )
            _delta = np.max([_delta, np.abs(temp_val[i] - old_v)])

        if _delta < delta:
            break

    return temp_val
```

```python
def policy_potential_evaluation(policy: Policy, dynamic: Dynamic) -> Values:
    # temp_val = [i for i in range(len(policy))]
    # temp_val = [1] * len(policy)
    temp_val = np.random.random(len(policy))

    while True:
        _delta = 0.0
        for (i, v) in enumerate(temp_val):
            old_v = v

            allowed_acts = acts(i)
            temp_val[i] = np.max(
                [
                    np.sum(
                        list(
                            map(
                                lambda newStateProbReward: newStateProbReward[1][0]
                                * (
                                    newStateProbReward[1][1]
                                    + gamma
                                    * (
                                        temp_val[newStateProbReward[0] - 1]
                                        if newStateProbReward[0] != -1
                                        else 0
                                    )
                                ),
                                dynamic(i + 1, act).items(),
                            )
                        )
                    )
                    for act in allowed_acts
                ]
            )
            _delta = np.max([_delta, np.abs(temp_val[i] - old_v)])

        if _delta < delta:
            break

    return temp_val
```

```python
def get_greedy_policy(values: Values, dynamic: Dynamic) -> Policy:
    temp_policy = [{} for _ in range(len(values))]
    for s in states:
        allowed_acts = acts(s)

        act_vals = [
            np.sum(
                list(
                    map(
                        lambda newStateProbReward: newStateProbReward[1][0]
                        * (
                            newStateProbReward[1][1]
                            + gamma
                            * (
                                (
                                    values[newStateProbReward[0] - 1]
                                    if newStateProbReward[0] != -1
                                    else 0
                                )
                            )
                        ),
                        dynamic(s + 1, act).items(),
                    )
                )
            )
            for act in allowed_acts
        ]
        new_act = allowed_acts[np.argmax(act_vals)]
        temp_policy[s][new_act] = 1.0

    return temp_policy
```

```python
def equiv_values(val1: Values, val2: Values) -> bool:
    if len(val1) != len(val2):
        return False

    for i in range(len(val1)):
        if val1[i] != val2[i]:
            return False

    return True


def equiv_policy(ply1: Policy, ply2: Policy) -> bool:
    if len(ply1) != len(ply2):
        return False
    for i in range(len(ply1)):
        acts1 = ply1[i]
        acts2 = ply2[i]
        if len(acts1.keys()) != len(acts2.keys()):
            return False

        if list(acts1.keys())[0] != list(acts2.keys())[0]:
            return False

    return True
```

```python
def iteration(
    init_val: Values,
    init_ply: Policy,
    dmc: Dynamic,
) -> Tuple[Values, Policy]:
    val_i = init_val
    ply_i = init_ply

    cnt = 0
    while True:
        print(f"in iteration {cnt}:")
        print(val_i)
        print(ply_i)
        new_vals = policy_evaluation(policy=ply_i, dynamic=dmc)

        new_policy = get_greedy_policy(values=new_vals, dynamic=dmc)

        if equiv_values(new_vals, val_i) and equiv_policy(new_policy, ply_i):
            print("converge, end!")
            return (new_vals, new_policy)
        elif cnt > 3_000:
            print("max iteration times exceed, end!")
            return (new_vals, new_policy)
        else:
            print(f"after iteration {cnt}:")
            print(new_vals)
            print(new_policy)
            val_i = new_vals
            ply_i = new_policy
            cnt += 1
```

```python
def values_print(vals: Values):
    pad_vals = [i for i in vals]
    pad_vals.append(0)
    pad_vals.insert(0, 0)
    for (i, val) in enumerate(pad_vals):
        print("{:5.1f}".format(val), end="")
        print(" ", end="")
        # print(str(val) + " ", end="")
        if i != 0 and (i + 1) % 4 == 0:
            print()


def policy_print(policy: Policy):
    pad_ply = [p for p in policy]
    pad_ply.append({"n": 0.0})
    pad_ply.insert(0, {"n": 0.0})
    for i, ply in enumerate(pad_ply):
        act = list(ply.keys())[0]
        if act == "l":
            token = "←"
        elif act == "u":
            token = "↑"
        elif act == "r":
            token = "→"
        elif act == "d":
            token = "↓"
        elif act == "n":
            token = " "
        else:
            raise ValueError(f"action {act} not expected")
        print(token + " ", end="")
        if (i + 1) % 4 == 0:
            print()
```
