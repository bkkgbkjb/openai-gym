```python
def policy(state: States) -> Dict[AllActions, float]:
    return {"r": 0.25, "l": 0.25, "d": 0.25, "u": 0.25}
```

```python
values = [float(i) for i in range(14)]
```

```python
def evaluation(
    val: List[float],
    policy: Callable[[States], Dict[AllActions, float]],
    dynamic: Callable[[States, AllActions], Dict[AllStates, Tuple[float, float]]],
    delta: float,
    gamma: float,
) -> List[float]:
    temp_val = [i for i in val]
    while True:
        _delta = 0.0
        for (i, v) in enumerate(temp_val):
            old_v = v
            temp_val[i] = np.sum(
                list(
                    map(
                        lambda actprob: actprob[1]
                        * np.sum(
                            list(
                                map(
                                    lambda newStateProbReward: newStateProbReward[1][0]
                                    * (
                                        newStateProbReward[1][1]
                                        + gamma
                                        * (
                                            temp_val[newStateProbReward[0] - 1]
                                            if newStateProbReward[0] != -1
                                            else 0
                                        )
                                    ),
                                    dynamic(i + 1, actprob[0]).items(),
                                )
                            )
                        ),
                        policy(i + 1).items(),
                    )
                )
            )
            _delta = np.max([_delta, np.absolute(temp_val[i] - old_v)])

        if _delta < delta:
            break

    return temp_val
```

```python
print(evaluation(values, policy, dynamics, 0.001, 1))
```
